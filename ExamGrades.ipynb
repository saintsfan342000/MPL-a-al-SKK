# Let's see how Wednesday 4-6 did on their midterm

```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as p
%matplotlib inline
p.style.use('mysty')
p.rcParams["figure.figsize"] = (12,12)
import figfun
import os
```

```python
d = pd.read_csv('Wed46.csv')

ex = d['Exam 2 (3957950)']

print('Mean: {:.1f}\nStDv: {:.1f}\nMax : {:.0f}\nMin : {:.0f}'.format(ex.mean(),ex.std(),ex.max(),ex.min()))
```

```python
ex.value_counts().sortlevel(ascending=False)
```

```python
ex.hist(bins=n.arange(50,110,10),rwidth=.8)
p.axis(ymax=15);
```

# Let's see how Wednesday 6-8 did

```python
d2 = pd.read_csv('Wed68.csv')

ex2 = d2['Exam 2 (3957976)']

print('Mean: {:.1f}\nStDv: {:.1f}\nMax : {:.0f}\nMin : {:.0f}'.format(ex2.mean(),ex2.std(),ex2.max(),ex2.min()))
```

```python
ex2.value_counts().sortlevel(ascending=False)
```

```python
ex2.hist(bins=n.arange(20,110,10),rwidth=.8,color='#5DA5DA')
p.axis(ymax=10);
```

```python
p.figure();
p.hist( [ex.values,ex2.values], bins=n.arange(50,110,10),range=(50,100), label=['4-6','6-8'] );
p.axis(ymax=15);
p.legend(loc='upper left');
```

```python
p.figure();
p.hist( [ex.values,ex2.values], bins=n.arange(50,110,10),range=(50,100), label=['4-6','6-8'], normed=True );
#p.axis(ymax=15);
p.legend(loc='upper left');
p.title('Normed');
```
