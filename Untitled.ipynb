```python
%matplotlib inline
import numpy as n
import matplotlib.pyplot as p
import figfun as f
p.style.use('mysty')
```

# 1018 Steel

```python
s = n.genfromtxt('Steel.dat')
```

```python
p.figure(figsize=(16,16))
p.plot(s[:,0],s[:,1],'b')
p.axis(xmin=0,ymin=0)
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Steel - Uniaxial Tension Test')
f.myax(p.gca())
```

```python
p.figure(figsize=(16,16))
p.plot(s[:,0],s[:,1],'b')
p.axis(xmin=0,ymin=0,xmax=.04)
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Linear and Perfectly Plastic Region')
f.myax(p.gca())
```

```python
p.figure(figsize=(16,16))
p.plot(s[:,0],s[:,1],'b')
p.axis(xmin=0,ymin=0,xmax=.04)
yloc = n.argmax(s[s[:,0]<.005,1])
p.plot(s[yloc,0],s[yloc,1],'ro',ms=10,label='Yield Point\n$\\epsilon$ = {:.4f}\n$\\sigma$ = {:.1f} ksi'.format(s[yloc,0],s[yloc,1]))
p.axis(xmin=0,ymin=0)
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Linear and Perfectly Plastic Region')
p.legend(loc='lower right',fontsize=20)
f.myax(p.gca())
```

```python
p.figure(figsize=(16,16))
p.plot(s[:,0],s[:,1],'b')
p.axis(xmin=0,ymin=0)
loc = n.argmax(s[:,1])
p.plot(s[yloc,0],s[yloc,1],'ro',ms=10,label='Yield Point\n$\\epsilon$ = {:.4f}\n$\\sigma$ = {:.1f} ksi'.format(s[yloc,0],s[yloc,1]))
p.plot(s[loc,0],s[loc,1],'r^',ms=10,label='Ultimate Stress\n$\\epsilon$ = {:.3f}\n$\\sigma$ = {:.0f} ksi'.format(s[loc,0],s[loc,1]))
p.plot(s[-1,0],s[-1,1],'rx',ms=10,mew=2,label='Failure\n$\\epsilon$ = {:.3f}\n$\\sigma$ = {:.0f} ksi'.format(s[-1,0],s[-1,1]))
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Steel - Uniaxial Tension Test')
p.legend(loc='lower right',fontsize=20)
f.myax(p.gca())
```

# Aluminum

```python
d = n.genfromtxt('Alu.dat')
```

```python
p.figure(figsize=(16,16))
p.plot(d[:,0],d[:,1],'b')
p.axis(xmin=0,ymin=0)
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Aluminum - Uniaxial Tension Test')
f.myax(p.gca())
```

```python
p.figure(figsize=(16,16))
p.plot(d[:,0],d[:,1],'b')
p.axis(xmin=0,ymin=0,xmax=.02)
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Uniaxial Tension Test')
f.myax(p.gca())
```

```python
rng = (d[:,0]<.003) 
E,b = n.polyfit(d[rng,0],d[rng,1],1)
p.figure(figsize=(16,16))
p.plot(d[:,0],d[:,1],'b')
x = n.linspace(.002,60/E+.002)
y = E*(x-.002)
p.plot(x,y)
p.plot(.00684786,50.4168,'ro',ms=10,label='Yield Point\n$\\epsilon$ = .0068\n$\\sigma$ = 50.4 ksi')
p.axis(xmin=0,ymin=0,xmax=.02)
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Uniaxial Tension Test')
p.legend(loc='lower right',fontsize=20)
f.myax(p.gca())
```

```python
p.figure(figsize=(16,16))
p.plot(d[:,0],d[:,1],'b')
p.axis(xmin=0,ymin=0)
loc = n.argmax(d[:,1])
p.plot(.00684786,50.4168,'ro',ms=10,label='Yield Point\n$\\epsilon$ = .0068\n$\\sigma$ = 50.4 ksi')
p.plot(d[loc,0],d[loc,1],'r^',ms=10,label='Ultimate Stress\n$\\epsilon$ = {:.3f}\n$\\sigma$ = {:.0f} ksi'.format(d[loc,0],d[loc,1]))
p.plot(d[-1,0],d[-1,1],'rx',ms=10,mew=2,label='Failure\n$\\epsilon$ = {:.3f}\n$\\sigma$ = {:.0f} ksi'.format(d[-1,0],d[-1,1]))
p.xlabel('$\\epsilon$',fontsize=30)
p.ylabel('$\\sigma$\n(ksi)',fontsize=30)
p.title('Steel - Uniaxial Tension Test')
p.legend(loc='lower right',fontsize=20)
f.myax(p.gca())
```
